## Example 2: If/Else — Even or Odd with if-else
**Concept:** if-else branching
**Explanation:** An `else` block runs when the `if` condition is false. Use `%` (modulo) to test divisibility by 2.


```python
number = 7
if number % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")  # 7 % 2 is 1, so else runs
```

## Example 3: If/Else — Letter Grade with elif
**Concept:** if-elif-else ladder
**Explanation:** `elif` (else-if) lets you test multiple conditions in order. The first match wins; later branches are skipped.


```python
score = 82
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"  # 82 lands here
elif score >= 70:
    grade = "C"
else:
    grade = "F"
print(f"Score {score} -> Grade {grade}")
```

## Example 91: While Loop — Binary Search with while
**Concept:** Halving search space
**Explanation:** Maintain low/high pointers until they cross.


```python
arr = [2, 5, 8, 12, 16, 23, 38]
target = 16
lo, hi = 0, len(arr) - 1
while lo <= hi:
    mid = (lo + hi) // 2
    if arr[mid] == target:
        print(f"Found at index {mid}")
        break
    elif arr[mid] < target:
        lo = mid + 1
    else:
        hi = mid - 1
```
